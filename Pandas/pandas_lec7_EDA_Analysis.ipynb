```python
'''
Higher the Features less will be the ML Prediction Score
# to convert categorical column into numerical -> use Encoders
  2 types :
    -> ordinal encoder
    -> label encoder
'''
```

```python
categorical_data=df.select_dtypes(include='object')
categorical_data
```

```python
numerical_data=df.select_dtypes(exclude='object')
numerical_data
```

```python
from collections import OrderedDict

num_stats=[]


for i  in numerical_data.columns :
  numerical_stat=OrderedDict({
      'Feature':i,
      'Count':numerical_data[i].count(),
      'Max':numerical_data[i].max(),
      'Min':numerical_data[i].min(),
      'Mean':numerical_data[i].mean(),
      'Median':numerical_data[i].median(),
      'Q1':numerical_data[i].quantile(0.25),
      'Q3':numerical_data[i].quantile(0.75),
      'IQR':numerical_data[i].quantile(0.75)-numerical_data[i].quantile(0.25),
      'Skewness':numerical_data[i].skew(),
      'Kurtosis':numerical_data[i].kurtosis()
  })
  num_stats.append(numerical_stat)
num_report=pd.DataFrame(num_stats)
num_report
```
